```python
# Exerscise 1 a,b

from bokeh.plotting import figure, output_notebook, show

x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

output_notebook()

p = figure(title="Scatter plot", x_axis_label='x', y_axis_label='y')

p.diamond(x, y, size=20, color="gold")

show(p)
```

```python
# Exercise 1 c,d

from bokeh.plotting import figure, output_notebook, show

x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

output_notebook()

p = figure(title="Scatter plot", x_axis_label='x', y_axis_label='y')

p.diamond(x, y, size=20, color=["gold", "red", "blue", "pink", "green"])

p.line(x,y)

show(p)
```

```python
# Exercise 3

fruits = ['Apples', 'Pears', 'Nectarines', 'Plums', 'Grapes', 'Strawberries']
counts = [5, 3, 4, 2, 4, 6]

p = figure(y_range=fruits, plot_height=250, title="Fruit Counts")

p.hbar(y=fruits, right=counts, height=0.9)

p.x_range.start = 0

show(p)
```

```python
# Exercise 4

from bokeh.models import ColumnDataSource
from bokeh.sampledata.autompg import autompg as df
df.head()

source = ColumnDataSource(df)

p = figure(plot_width=400, plot_height=400, title="Autompg dataset", x_axis_label='Index', y_axis_label='Weight')
p.circle('index', 'weight', source=source)

show(p)
```

```python
# Exercise 5

from bokeh.layouts import column
from bokeh.models import Slider
from bokeh.plotting import figure, show

plot = figure(plot_width=400, plot_height=400)
r = plot.circle([1,2,3,4,5,], [3,2,5,6,4], radius=0.2, alpha=0.5)

slider = Slider(start=0.1, end=2, step=0.01, value=0.2, title="Glyph opacity")
slider.js_link('value', r.glyph, 'fill_alpha')

show(column(plot, slider))
```
